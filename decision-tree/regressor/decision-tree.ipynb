# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor , export_graphviz
from sklearn.metrics import mean_squared_error

from six import StringIO
```

```python
X, y = make_regression(n_features=4, random_state=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8, random_state=0)
```

```python
# definition
model = DecisionTreeRegressor(random_state=0)

# training
model.fit( X_train, y_train )

# performance
y_pred = model.predict( X_test )
```

```python
mse = mean_squared_error( y_test, y_pred)
rmse = np.sqrt(mse)
```

```python
# Export

dor_data = StringIO

export_graphviz(
    model,
    out_file='tree_reg.dot',
    rounded=True,
    filled=True
    )
```

```python
# definition
model = DecisionTreeRegressor(max_depth=2,random_state=0)

# training
model.fit( X_train, y_train )

# performance
y_pred = model.predict( X_test )

# metrics
mse = mean_squared_error( y_test, y_pred)
rmse = np.sqrt(mse)

print(f'RMSE {round(rmse,3)}, MSE:{round(mse,3)}')

# export
dor_data = StringIO

export_graphviz(
    model,
    out_file='tree_reg.dot',
    rounded=True,
    filled=True
    )

!dot -Tpng tree_reg.dot -o tree_reg.png

img = cv2.imread('tree_reg.png')
plt.figure( figsize= (20,20) )
plt.imshow(img)
```

```python
max_depth_values = np.arange(2,25,1)
rmse_list = []
for i in max_depth_values:
    # definition
    model = DecisionTreeRegressor(max_depth=i,random_state=0)

    # training
    model.fit( X_train, y_train )

    # performance
    y_pred = model.predict( X_test )

    # metrics
    mse = mean_squared_error( y_test, y_pred)
    rmse = np.sqrt(mse)
    rmse_list.append( rmse )
```

```python
plt.figure( figsize= (10,10) )
plt.plot(max_depth_values, rmse_list, marker='o')
plt.ylabel('rmse')
plt.xlabel('max_depths')
plt.ylim( [0,120] )
```

```python
X, y = make_regression(n_samples=20000,n_features=4, random_state=0)
# Train test Split
X, X_test, y, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
# Train val Split
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.2, random_state=0)

max_depth_values2 = np.arange(2,30,1)
rmse2_list = []

for i in max_depth_values2:
    # definition
    model = DecisionTreeRegressor(max_depth=i,random_state=0)

    # training
    model.fit( X_train, y_train )

    # performance
    y_pred = model.predict( X_val )

    # metrics
    mse = mean_squared_error( y_val, y_pred)
    rmse = np.sqrt(mse)
    rmse2_list.append( rmse )
```

```python
plt.figure( figsize= (10,10) )
plt.plot(max_depth_values2, rmse2_list, marker='o')
plt.ylabel('rmse')
plt.xlabel('max_depths')
plt.ylim( [10,75] )
```

```python
min_samples_split_values = np.arange(2,30,1)
rmse2_list = []

for i in min_samples_split_values:
    # definition
    model = DecisionTreeRegressor(min_samples_split=i,random_state=0)

    # training
    model.fit( X_train, y_train )

    # performance
    y_pred = model.predict( X_val )

    # metrics
    mse = mean_squared_error( y_val, y_pred)
    rmse = np.sqrt(mse)
    rmse2_list.append( rmse )
plt.figure( figsize= (10,10) )
plt.plot(min_samples_split_values, rmse2_list, marker='o')
plt.ylabel('rmse')
plt.xlabel('min_samples_split')
```
