```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve, f1_score, auc
from scikitplot.metrics import plot_precision_recall_curve, plot_lift_curve, plot_cumulative_gain, plot_roc_curve, plot_ks_statistic
from matplotlib import pyplot as plt
```

```python
X, y = make_classification(
                            n_samples=1000, 
                            n_classes=2, 
                            random_state=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=2)
```

```python
model = DecisionTreeClassifier(max_depth=2)

model.fit(X_train,y_train)

y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)
y_hat_test = model.predict_proba(X_test)
y_hat_test = y_hat_test[:,1]
```

```python
precision,recall,treshhold = precision_recall_curve(y_test,y_hat_test)
```

```python
plot_precision_recall_curve(y_test,y_pred_proba)
```

```python
plt.plot(recall,precision,marker='.',label='model')
plt.xlabel( 'Recall' )
plt.xlabel( 'Precision' )
```

```python
model2 = LogisticRegression(solver='lbfgs')
model2.fit(X_train,y_train)
y_pred2 = model2.predict(X_test)
y_pred_proba2 = model2.predict_proba(X_test)
y_hat_test2 = model2.predict_proba(X_test)
y_hat_test2 = y_hat_test2[:,1]
```

```python
precision2,recall2,treshhold2 = precision_recall_curve(y_test,y_hat_test2)
```

```python
plot_precision_recall_curve(y_test,y_pred_proba2)
```

```python
plt.plot(recall2,precision2,marker='.',label='model2')
plt.xlabel( 'Recall' )
plt.xlabel( 'Precision' )
```

```python
X, y = make_classification(
                            n_samples=500, 
                            n_classes=2, 
                            random_state=2,
                            n_redundant=2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=2)
```

```python
model2 = LogisticRegression(solver='lbfgs')
model2.fit(X_train,y_train)
y_pred2 = model2.predict(X_test)
y_pred_proba2 = model2.predict_proba(X_test)
y_hat_test2 = model2.predict_proba(X_test)
y_hat_test2 = y_hat_test2[:,1]
```

```python
precision2,recall2,treshhold2 = precision_recall_curve(y_test,y_hat_test2)
```

```python
plot_precision_recall_curve(y_test,y_pred_proba2)
```

```python
plt.figure(figsize=(10,10))
plt.plot(recall2,precision2,marker='o',label='model2')
plt.xlabel( 'Recall' )
plt.xlabel( 'Precision' )
```
