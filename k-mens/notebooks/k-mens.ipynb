# 0.0 Imports

```python
from sklearn import datasets as ds
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics as mt
import numpy as np
```

# 1.0 Loading Dataset

```python
X,y = ds.make_blobs(
    n_samples=100,
    n_features=2,
    centers=4,
    cluster_std=0.9,
    random_state=0
)
```

```python
plt.scatter(
        X[:,0]
        ,X[:,1]
        ,c=y
        )
```

# 2.0 Training Model

```python
n_clusters=2

# Define
Kmeans = KMeans(
                n_clusters=n_clusters
                ,init='random'
                ,n_init=10
                ,random_state=0
                )

#Fit
Kmeans.fit( X )
labels = Kmeans.fit_predict( X )
# Performance ( test)
labels = Kmeans.fit_predict( X=X )
silhouete_avg = mt.silhouette_score( X,labels )

print(f'Para n_clusters = {n_clusters}' ,
      f'A silueta Média é {silhouete_avg}')
```

```python
plt.scatter(
        X[:,0]
        ,X[:,1]
        ,c=Kmeans.labels_
        )
```

# 3.0 Fine Tuning

```python
n_clusters=np.arange(2,11,1)

ss_list = []
for c in n_clusters:
    Kmeans = KMeans(
                    n_clusters=c
                    ,init='random'
                    ,n_init=10
                    ,random_state=0
                    )

    #Fit
    Kmeans.fit( X )
    labels = Kmeans.fit_predict( X )
    # Performance ( test)
    labels = Kmeans.fit_predict( X=X )
    ss_avg = mt.silhouette_score( X,labels )
    ss_list.append(ss_avg)

    print(f'Para n_clusters = {c}' ,
        f'A silueta Média é {ss_avg}')
```

```python
plt.plot(
        n_clusters
        ,ss_list
        ,marker='o')
plt.xlabel('Numero de clusters K')
plt.ylabel('Score Silhueta Média')
```

# 5.0 Avaliando Modelos de Cluster

## Coesão

A coesão mede o quanto os pontos em um cluster estão próximos uns dos outros.
Ela é medida pela soma do erro das distâncias quadráticas entre cada ponto do cluster e o seu
centroide dentro do próprio cluster. Essa medida de coesão é conhecida com "Within-Cluster Sum
of Squares", do inglês, WCSS

### $WCSS = {\sum_{i=1}^n{ (x_i - c_j)^2}}$


## Separação

A separação mede o quão distante os pontos em um cluster estão de outros clusters. Ela é
medida pela distância entre os centróides dos clusters. Quanto maior a distância entre os
centróides, mais separados os clusters estão.

Em outras palavras, a separação mede o quão bem definidos os clusters estão uns dos outros.
Um baixo valor de separação indica que os clusters podem estar muito próximos uns dos outros 
o que pode tornar difícil a sua distinção. Por outro lado, um valor alto de separação sugere que os
clusters são bem definidos e distintos uns dos outros.

## Silhouette Score

O Silhouette Score é uma medida de avaliação para medir a qualidade de um agrupamento. Ele é
calculado para cada observação e mede o quão bem ela se encaixa no seu cluster atual, em
comparação com outros clusters.

O Silhouette Score varia de -1 a 1, onde um valor mais próximo de 1 indica que a observação
está bem ajustada ao seu cluster e mal ajustada aos outros clusters, e um valor mais próximo de
-1 indica que a observação está mal ajustada ao seu cluster e bem ajustada aos outros clusters.

Um valor próximo de 0 indica que a observação está entre dois clusters e pode ser atribuída a
qualquer um deles

### $si = \frac{b_i - a_I}{max(a_i,b_i)}$

# 6.0 Elbow Method (Método de Cotovelo)

### Como funciona o Elbow Method?

O Elbow Method é um método que ajuda a determinar o número ideal de
clusters (ou grupos) em uma análise de clusterização. Ele é chamado de Elbow
Method (método do cotovelo) porque a curva formada pelo número de clusters
em relação à variabilidade total dos dados se assemelha a um braço dobrado no
cotovelo.

### Os 4 passos de funcionamento do Elbow Method

1. Realize a clusterização dos dados para diferentes valores de K.
2. Para cada valor de K, calcule a soma dos quadrados das distâncias de cada
ponto ao centro do cluster mais próximo ( Coesão - WCSS ) - SS
3. Trace um gráfico com o número de clusters no eixo x e a soma das
distâncias no eixo y.
4. O objetivo é escolher o valor de K que resulte em uma redução significativa
da soma das distâncias em relação ao número de clusters. Isso é indicado
por uma curva que se assemelha a um braço dobrado em um cotovelo, onde
o ponto de dobra indica o valor ideal de K.

# 7.0 Elbow Method  Prática

```python
import sklearn.datasets     as ds
import matplotlib.pyplot    as plt
import sklearn.cluster      as ct
import sklearn.metrics      as mt
import numpy                as np
```

```python
X,y =ds.make_blobs(
                    n_samples=100,
                    n_features=4,
                    centers=3,
                    cluster_std=0.3,
                    random_state=0
                    )
```

```python
clusters = np.arange(2,11,1)
ss_list = []

for k in clusters:  
    model = ct.KMeans( n_clusters=k,init='random',n_init=10,random_state=0)
    labels = model.fit_predict( X )
    ss = mt.silhouette_score( X, labels)
    ss_list.append( ss )
    print(f'clusters K: {k} SS: {ss}')

plt.plot(clusters,ss_list,marker='o',linewidth=2)
plt.xlabel( 'K value' )
plt.ylabel( 'Silhouette AVG' )
```
